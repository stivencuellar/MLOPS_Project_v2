```python
#Tabla de frecuencias para total_bill con intervalos automáticos

def tabla_frecuencia_total_bill(datos):
    columna = datos['total_bill']  # Seleccionar la columna de interés
    n = len(columna)  # Cantidad de datos
    k = math.ceil(1 + math.log2(n))  # Regla de Sturges
    min_val, max_val = min(columna), max(columna)
    rango = max_val - min_val
    amplitud = rango / k  # Amplitud de los intervalos

    # Definir los intervalos
    bins = np.linspace(min_val, max_val, k + 1)

    # Calcular frecuencias
    frecuencias, _ = np.histogram(columna, bins)

    # Calcular frecuencias acumuladas
    frec_acumulada = np.cumsum(frecuencias)

    # Calcular frecuencia relativa
    frec_relativa = frecuencias / n

    # Calcular frecuencia relativa acumulada
    frec_rel_acumulada = np.cumsum(frec_relativa)

    # Crear tabla de frecuencias
    tabla = pd.DataFrame({
        'Intervalo': [f'[{bins[i]:.2f}, {bins[i+1]:.2f})' for i in range(len(bins)-1)],
        'Frecuencia': frecuencias,
        'Frec. Acumulada': frec_acumulada,
        'Frec. Relativa': frec_relativa,
        'Frec. Rel. Acumulada': frec_rel_acumulada
    })

    return tabla

resultado_total_bill = tabla_frecuencia_total_bill(datos)
print(resultado_total_bill)
```

- Distribución no uniforme: los datos no se distribuyen homogéneamente. Casi el 57% de las facturas están por debajo de $19.00 , lo cual indica una alta concentración en consumos bajos a medios .

- Asimetría a la derecha (positiva): se observa que los primeros tres intervalos concentran la mayoría de las observaciones (73 + 59 + 8 = 140 casos ≈ 57%). Los intervalos finales tienen muy poca frecuencia → colas largas hacia la derecha , lo cual confirma el valor de asimetría de +1.13 encontrado previamente.

- Clientes de alto gasto son minoría: solo el 3.6% de los clientes pagan facturas superiores a $40.20 (últimos dos intervalos), lo que refuerza la presencia de valores atípicos de alto valor .
